```python
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib

%matplotlib inline
```

```python
image = imageio.imread('dog.4036.jpg')
ia.imshow(image)
```

```python
image.shape
```

```python
rotate = iaa.Affine(rotate = (-180, 0))
rotated_image = rotate.augment_image(image)
ia.imshow(rotated_image)
```

```python
noise = iaa.AdditiveGaussianNoise(10, 80)
noise_image = noise.augment_image(image)
ia.imshow(noise_image)
```

```python
crop = iaa.Crop(percent = (0, 0.4))
cropped_image = crop.augment_image(image)
ia.imshow(cropped_image)
```

```python
shear = iaa.Affine(shear = (0, 80))
sheared_image = shear.augment_image(image)
ia.imshow(sheared_image)
```

```python
flip_hr = iaa.Fliplr(p = 1.0)
flipped_image = flip_hr.augment_image(image)
ia.imshow(flipped_image)
```

```python
flip_vr = iaa.Flipud(p = 1.0)
flipped_image = flip_vr.augment_image(image)
ia.imshow(flipped_image)
```

```python
contrast = iaa.GammaContrast(gamma = 0.1)
contrasted_image = contrast.augment_image(image)
ia.imshow(contrasted_image)
```

```python
scale = iaa.Affine(scale={"x": (2.0, 1.0), "y": (1.5, 1.0)})
scaled_image = scale.augment_image(image)
ia.imshow(scaled_image)
```
